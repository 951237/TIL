```
import pandas as pd
```

```
col = ['a', 'b', 'c']
ind = [1, 2, 3]
con = [[0,0,0], [0,0,0], [0,0,0]]
pd.DataFrame(con, columns=col, index=ind) # 데이터 프레임 만들기 내용, 칼럼, 인덱스 
```

```
def df_maker(col_num, ind_num, fill):
    col = []
    ind = []
    con = []
    for i in range(0, col_num):
        col.append(fill)
    for i in range(0, ind_num):
        ind.append(fill)
    for i in range(0, ind_num):
        con.append(col)

    return pd.DataFrame(con, columns=col, index=ind)
```

```
df_maker(15,15,"Nan")
```
