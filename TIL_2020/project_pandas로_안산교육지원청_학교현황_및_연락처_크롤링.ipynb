### 안산교육지원청 유치원 현황

```
import pandas as pd
from google.colab import files

url = 'http://www.goeas.kr/USR/ORG/MNU8/SchoolList.do?orgType=A'
df = pd.read_html(url, header = 0, index_col = 0) # 헤더를 첫행으로, 인덱스는 0번째 열로 정하기
kinder = df[0]
kinder.to_excel('/content/drive/My Drive/data/result/안산_유치원현황_{}.xlsx'.format(timestr))
files.download('/content/drive/My Drive/data/result/안산_유치원현황_{}.xlsx'.format(timestr))
# 데이터 프레임 첫행 삭제하기
# 원하는 유치원 정보만 살펴보기
```
