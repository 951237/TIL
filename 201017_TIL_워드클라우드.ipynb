```
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```
PATH_TXTFILE = "/content/example.txt"
PATH_IMG = "/content/drive/My Drive/data/워드클라우드/study_이상한 나라 앨리스/09. alice_mask.png"
```

```
text = open(PATH_TXTFILE, encoding='utf-8').read()
alice_mask = np.array(Image.open(PATH_IMG))
```

```
# 많이 나오는 단어 제거
stopwords = set(STOPWORDS)
stopwords.add("said")
```

```
# 코랩 한글 설정하기
%matplotlib inline  

import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도

!apt-get update -qq
!apt-get install fonts-nanum* -qq

path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'  # 설치된 나눔글꼴중 원하는 녀석의 전체 경로를 가져오자
font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)
plt.rc('font', family=font_name)



# 우선 fm._rebuild() 를 해주고
fm._rebuild()

#폰트 크기 변경
plt.rcParams.update({'font.size': 20})
```

```
# 마스크 이미지 확인하기
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
# 텍스트 워드클라우드 분석하기
wc = WordCloud(font_path = path, background_color='white', max_words=2000, mask=alice_mask, 
               stopwords = stopwords) # 한글 폰트 경로 설정하기
wc = wc.generate(text)
wc.words_
```

```
# 워드클라우드 이미지 만들기
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
