```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
PATH_TEST = 'https://raw.githubusercontent.com/corazzon/KaggleStruggle/master/bike-sharing-demand/data/test.csv'
PATH_TRAIN = 'https://raw.githubusercontent.com/corazzon/KaggleStruggle/master/bike-sharing-demand/data/train.csv'
```

```python
train = pd.read_csv(PATH_TRAIN, parse_dates=['datetime'])
test = pd.read_csv(PATH_TEST, parse_dates=['datetime'])
```

### feature engineering

```python
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour
```

```python
test['year'] = test['datetime'].dt.year
test['month'] = test['datetime'].dt.month
test['day'] = test['datetime'].dt.day
test['hour'] = test['datetime'].dt.hour
```

```python
# widspeed 풍속에 0이 가장 많음. → 잘못 기록된 데이터 고치기
fig, axes = plt.subplots(nrows=2)
fig.set_size_inches(18, 10)

plt.sca(axes[0])
plt.xticks(rotation=30, ha = 'right')
axes[0].set(ylabel = 'Count', title="train windspeed")
sns.countplot(data=train, x='windspeed', ax = axes[0])

plt.sca(axes[1])
plt.xticks(rotation=30, ha = 'right')
axes[1].set(ylabel = 'Count', title="test windspeed")
sns.countplot(data=test, x='windspeed', ax = axes[1])
```
