```python
# 그래프 라이브러리 호출
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic') 

# plt.figure(figsize=(6,16))
fig, ax = plt.subplots(1,3, figsize=(16,4))

sns.lineplot(x="year", y = 'avgPrice', data = df_cabbage, ax = ax[0])
sns.lineplot(x='year', y = 'avgTemp', data = df_cabbage, ax = ax[1])
sns.lineplot(x="year", y = 'rainFall', data = df_cabbage, ax = ax[2])
ax[0].set_title('Price')
ax[1].set_title('Temp')
ax[2].set_title('rainFall')
plt.suptitle('Main')
plt.show()
```
