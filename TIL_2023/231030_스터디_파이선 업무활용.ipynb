## 231029_텍스트 처리
- 텍스트 파일 열기
- 텍스트 파일 읽기
- 텍스트 파일에서 '일시', '온도'만 추출하기
- 텍스트 파일 다른 이름으로 저장하기

```python
# 텍스트 파일 열기
file_name = './data/WeatherData_230301'
f = open(f"{file_name}.txt",'r')
```

```python
# 텍스트 파일 읽기
data = f.read()
```

```python
f.close()
```

```python
# 텍스트 파일 출력
print(data)
```

```python
data
```

```python
# 텍스트 파일 줄단위로 분할하기 - 리스트 저장
data = data.split('\n')
```

```python
data = data[1:]
data
```

```python
# 데이터 추출하기(시간, 온도)
string = "시간,온도\n"

for i in data:
    time = i.split(',')[0]
    temp = i.split(',')[1]
    string += f"{time},{temp}\n"
string = string[:-1]
print(string)
```

```python
f = open(f"{file_name}_modifi.txt",'w')
f.write(string)
f.close
```
