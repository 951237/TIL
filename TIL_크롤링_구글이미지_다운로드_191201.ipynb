### 라이브러리 불러오기

```
import urllib.request
from urllib.request import urlopen
from urllib.parse import quote_plus
from google.colab import files
import pandas as pd
from bs4 import BeautifulSoup
```

### url가져오기

```
path_downlad = '/content/drive/My Drive/data/result/img/'
baseurl = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
v_input = input('주인님, 검색어를 입력하세요 : ')
url = baseurl + quote_plus(v_input)
print(url)
```

```
html = urllib.request.urlopen(url).read()
print(html)
```

```
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```
imgs = soup.find_all(class_='_img')
print(imgs[0])

'''
<img alt="다이어트에 도움되는 식품 | 포스트" class="_img" data-height="250" data-source="https://search.pstatic.net/common/?src=http%3A%2F%2Fpost.phinf.naver.net%2F20150128_84%2Fhikanie_1422447546230gA9Cp_JPEG%2Fmug_obj_142244756790945395.jpg&amp;type=b400" data-width="250" onerror="var we=$Element(this); we.addClass('bg_nimg'); we.attr('alt','이미지준비중'); we.attr('src','data:image/gif;base64,R0lGODlhAQABAIAAAAAAAP///yH5BAEAAAAALAAAAAABAAEAAAIBRAA7');" src="data:image/gif;base64,R0lGODlhAQABAIAAAAAAAP///yH5BAEAAAAALAAAAAABAAEAAAIBRAA7"/>
'''
```

```
n = 1
for img in imgs:
    img_url = img['data-source']
    with urlopen(img_url) as f:
        with open(path_downlad + v_input + str(n) + '.jpg', 'wb') as h:
            img_file = f.read()
            h.write(img_file)
            print('{}/{}째 파일 다운로드 완료'.format(n, len(imgs)))
            
    n += 1

print('다운로드 완료')
```
